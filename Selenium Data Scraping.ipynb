```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
import time
import pandas as pd
```

Andhra Buses

```python
driver=webdriver.Chrome()

driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")

time.sleep(5)

#driver.maximize_window()
```

```python
wait = WebDriverWait(driver, 20)

def andhra_links_and_routes(x):
    Andhra_Link = []
    Andhra_Route = []
    
    for i in range(1,6):
        y = driver.find_elements(By.XPATH,x)

        for links in y:
            a = links.get_attribute('href')
            Andhra_Link.append(a)
    
        for route in y:
            Andhra_Route.append(route.text)
    
        try:
            pagination = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="DC_117_paginationTable"]')))
            next = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            time.sleep(3)
            next.click()
    
        except NoSuchElementException:
            print(f'No more data left in step {i}')
            break
            
    return Andhra_Link,Andhra_Route

Andhra_Link,Andhra_Route = andhra_links_and_routes("//a[@class='route']")
```

```python
df_Andhra=pd.DataFrame({"Route_name":Andhra_Route,"Route_link":Andhra_Link})
df_Andhra
```

```python
path=r"C:\Users\bhara\Project\df_Andhra.csv"
df_Andhra.to_csv(path,index=False)
```
