```python
import mysql.connector
import pandas as pd
import numpy as np
```

```python
df_all_buses = pd.read_csv('df_all_buses.csv')
df_all_buses
```

```python
df_all_buses.info()
```

```python
df_all_buses["Price"]=df_all_buses["Price"].str.replace("INR","")
df_all_buses["Price"]=df_all_buses["Price"].astype(float)
df_all_buses["Price"].fillna(0)
```

```python
df_all_buses["Ratings"]=df_all_buses["Ratings"].str.replace("New","")
df_all_buses["Ratings"]=df_all_buses["Ratings"].str.strip()
df_all_buses["Ratings"]=df_all_buses["Ratings"].str.split().str[0]
df_all_buses["Ratings"] = pd.to_numeric(df_all_buses["Ratings"], errors='coerce')
df_all_buses["Ratings"].fillna(0,inplace=True) 
```
